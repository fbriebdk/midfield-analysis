```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV in the current directory
df = pd.read_csv("data\player_stats.csv", encoding='latin1')

# Set player names as index
df.set_index("Player", inplace=True)

# Choose a few key metrics to visualize
key_metrics = ["xAG", "Assists", "SCA", "Prog Passes", "npxG + xAG"]

# Plot a grouped bar chart
df[key_metrics].plot(kind="bar", figsize=(12, 6))
plt.title("Key Creative Metrics per 90")
plt.ylabel("Per 90 Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
